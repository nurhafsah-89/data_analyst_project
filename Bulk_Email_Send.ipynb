```python
import openpyxl
from openpyxl import load_workbook
from email.mime.text import MIMEText
from smtplib import SMTP_SSL
from email.message import EmailMessage

# Email sending server
host_server = 'smtp.gmail.com'
sender_mail = 'mohdazhar4379@gmail.com'
sender_passcode = 'pkpcxlmwggjxrtao'

def send_mail(receiver='', mail_title='', mail_content='', files=None):
    # SSL login
    smtp = SMTP_SSL(host_server, 465)  # 465 587 993
    smtp.ehlo()
    smtp.login(sender_mail, sender_passcode)

    # Construct message
    msg = EmailMessage()
    msg.set_content(mail_content)
    msg["Subject"] = mail_title
    msg["From"] = sender_mail
    msg["To"] = receiver.strip().replace("\n", "").replace("\r", "")

    for file in files:
        with open(file, 'rb') as fp:
            data = fp.read()
        msg.add_attachment(data, maintype='application', subtype='octet-stream', filename=file)

    try:
        smtp.send_message(msg)
        print(f"Email sent successfully to {receiver}")
    except Exception as e:
        print(f"Error sending email to {receiver}: {e}")

    smtp.quit()


# Loading excel file
wb = load_workbook("database.xlsx")
sheet = wb[wb.sheetnames[0]]

# Loop over the excel file's rows
for rownum in range(2, sheet.max_row + 1):  # Assuming row 1 is the header row
    # Parse company name
    Company = sheet.cell(row=rownum, column=1).value
    # Parse contact name
    Contact_name = sheet.cell(row=rownum, column=2).value
    # Parse position
    Position = sheet.cell(row=rownum, column=3).value
    # Email
    Email = sheet.cell(row=rownum, column=4).value

    # Mail contents
    mail_content = f"Good Day {Contact_name},\n\nI am writing to express my sincere interest in the {Position} position at {Company}. With over 11 years of work experience in various roles, including as a Senior Sub-Business Coordinator at SMIN Vision Venture, I have developed a strong foundation in procurement and coordination, making me an ideal candidate for this position.\n\n I have attached my resume, which includes a detailed cover letter outlining my qualifications and achievements. It would be greatly appreciated if you could take a moment to review it and consider my candidacy.\n\nI believe a quick call would provide us with an opportunity to discuss how my skills and experiences align with the needs of your company. You can also find additional information about my background on my LinkedIn profile.\n\nI eagerly await your response and look forward to hearing from you soon. Thank you for considering my application.\n\nBest Regards\nMohamed Azhar\n+6017-4379369\nhttps://www.linkedin.com/in/mohamed-azhar-abdul-rahim/"
    # Mail title
    mail_title = 'Experienced Professional Seeking Buyer/Planner Opportunities'

    # Print the extracted values
    print(f"Company: {Company}")
    print(f"Contact Name: {Contact_name}")
    print(f"Position: {Position}")
    print(f"Email: {Email}")
    print()

    # Send email using the extracted email address
    send_mail(receiver=Email, mail_title=mail_title, mail_content=mail_content, files=["Resume_Azhar.pdf", "Cover_letter.pdf"])
```
